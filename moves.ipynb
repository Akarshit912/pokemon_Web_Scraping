```python
values=[]
n=len(soup.find_all('tr'))
```

```python

rows = soup.find_all('tr')

def safe_text(cell):
    return cell.get_text(strip=True) if cell else ""   # default placeholder

def safe_img_alt(cell):
    img = cell.find("img") if cell else None
    return img.get("alt") if img and img.has_attr("alt") else ""

for i in range(1, n):
    tds = rows[i].find_all("td")

    # ensure we can access safely
    def td(idx):
        return tds[idx] if len(tds) > idx else None

    values.append([safe_text(td(0)),
        safe_text(td(1)),safe_img_alt(td(2)),safe_text(td(3)),safe_text(td(4)),safe_text(td(5)),safe_text(td(6)) or " ", 
        safe_text(td(7)),
    ])

    print("done row:", i)
```

```python
import pandas as pd

df = pd.DataFrame(values, columns=['Name','Type','Cat','Power','Acc','PP','Effect','Prob'])
df.to_csv("moves.csv", index=False, encoding="utf-8")

print("CSV created!")
```
