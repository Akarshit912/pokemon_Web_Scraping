```python
values=[]
for j in range(4,6):
    pokemon=requests.get(url.format(j),headers=headers).text
    soup=BeautifulSoup(pokemon,'lxml')
    name=soup.find('h1').text.split()[0]
    n=len(soup.find('div',class_='resp-scroll').find_all('tr'))
    for i in range(1,n):
        a=soup.find('div',class_='resp-scroll').find_all('tr')[i].find_all('td')
        values.append([name,a[0].text,a[1].text,a[2].text])
```
