```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
url='https://pokemondb.net/pokedex/all'
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
pokemon=requests.get(url.format(6),headers=headers).text
```

```python
soup=BeautifulSoup(pokemon,'lxml')
```

```python
tag=soup.find_all('tr')[121].find_all('td')[0].find("source")
avif_link = tag.get("srcset", None)
img = tag.find("img")
png_link = img.get("src") if img else None

print(avif_link)
print(png_link)
```

```python
avif=[]
png=[]
index=[]
```

```python
for i in range (1,len(soup.find_all('tr'))):
    tag = soup.find_all('tr')[i].find_all('td')[0].find("source")

    avif_link = tag.get("srcset") if tag and tag.has_attr("srcset") else None
    
    img = tag.find("img") if tag else None
    png_link = img.get("src") if img and img.has_attr("src") else None
    index_val=soup.find_all('tr')[i].find_all('td')[0].find('span').text
    avif.append(avif_link)
    png.append(png_link)
    index.append(index_val)
```

```python
# soup.find_all('tr')[i].find_all('td')[0].find('span').text
```

```python
df = pd.DataFrame({
    "Index": index,
    "AVIF": avif,
    "PNG": png
})

df.to_csv("output1.csv", index=False)
```
