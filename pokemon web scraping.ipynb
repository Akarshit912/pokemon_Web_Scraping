```python
index=[]
list_pokemon=[]
species=[]
hieght=[]
weight=[]
abilities=[]
hidden_abilities=[]
catch_rate=[]
percet_at_full_HP=[]
friendship=[]
egg=[]
male=[]
type=[]
```

```python
for i in range(1,10):
    pokemon=requests.get(url.format(i),headers=headers).text
    soup=BeautifulSoup(pokemon,'lxml')
    
    name_val=soup.find_all('h1')[0].text
    
    index_val=soup.find_all('strong')[0].text
    
    species_val=soup.find_all('td')[2].text
    
    hieght_val = soup.find_all('td')[3].text.strip()  
    hieght_val = float(hieght_val.split('\xa0')[0])
    
    weight_val = soup.find_all('td')[4].text.strip()  
    weight_val= float(weight_val.split('\xa0')[0])

    abilities_val=soup.find_all('td')[5].find_all('a')[0].text
    hidden_abilities_val=soup.find_all('td')[5].find_all('a')[1].text

    catch=soup.find_all('td')[8].text
    catch = catch.strip()
    
    catch_rate_val = int(catch.split()[0])
    percent_val = float(catch.split('(')[1].split('%')[0])

    friend = soup.find_all('td')[9].text.strip()
    friendship_val = int(friend.split()[0])

    egg_val=soup.find_all('td')[12].text

    male_val = soup.find_all('td')[13].text
    male_val = float(male_val.split('%')[0])

    p_type=[]
    number=len(soup.find_all('td')[1].find_all("a"))
    for j in range(0,number):
        p_type.append(soup.find_all('td')[1].find_all("a")[j].text)
    
    list_pokemon.append(name_val)
    index.append(index_val)
    species.append(species_val)
    hieght.append(hieght_val)
    weight.append(weight_val)
    abilities.append(abilities_val)
    hidden_abilities.append(hidden_abilities_val)
    catch_rate.append(catch_rate_val)
    percet_at_full_HP.append(percent_val)
    friendship.append(friendship_val)
    egg.append([egg_val])
    male.append(male_val)
    type.append(p_type)
```

```python
print(list_pokemon,index,type,species,hieght,weight,abilities,hidden_abilities,catch_rate,percet_at_full_HP,friendship,male,egg,type)
```
